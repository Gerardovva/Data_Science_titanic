**Calcular la tarifa media de los pasajeros del titanic y crea una nueva columna llamada "Fare/mean Fare" donde calcules el ratio entre la tarifa que ha pagado cada pasajero entre la tarifa media** 

```python
media=df["Fare"].mean()
media
```

```python
df["Fare/mean_Fare"]=df["Fare"]/media
df.head(10)
```

**Muestra el precio mas elevado y el precio mas bajo de un ticket para le titanic. Tambien muestra la recaudacion total del titanic por venta de los ticket**

```python
bajo=df["Fare"].min()
alto=df["Fare"].max()
suma=df["Fare"].sum()
```

```python
print(f"precio mas bajo: {bajo} dolares")
print(f"precio mas alto: {alto} dolares")
print(f"total de venta: {suma} dolares")
```
